# MSA simulation and quantify gene tree estimation error (GTEE)

Here we follow the analyses done in [**Cai et al. (2021)**](https://doi.org/10.1093/sysbio/syaa083) to quantify the GTEE in our dataset.

Step 1: simulate MSA based on your original MSA and best-fit model. The command line is as below.

````bash
mkdir CoBi-GTEE;cd CoBi-GTEE
````

````bash
iqtree3 -s ~/CoBi-course/full_dataset_concat/concatenated.msa -m LG+F+R4 --alisim concatenated.msa.alignment_mimic --out-format fasta -t ~/CoBi-course/full_dataset_concat/concatenated.msa.treefile --num-alignments 100 -pre concatenated.msa.seqsimu -T 6
````

Step 2: calculate sCF for each simulated MSA. Below I give one example. You can submit jobs to `HPC` to finish the entire 100 MSAs.

````bash
iqtree3 -s concatenated.msa.alignment_mimic_100.fa -m LG+F+R4 -t ~/CoBi-course/full_dataset_concat/concatenated.msa.treefile --scfl 100 -pre concatenated.msa.alignment_mimic_100.fa.sCF
````

Here we will use the distribution of the sDF from these 100 simulated MSA to build the null distribution of sDF which lacks of `GTEE`.

````bash
python ~/CoBi-course/func/multifunction.py quantifygtee ~/CoBi-course/Simulated_MSA_sCF/*.stat -o GTEE.svg
````
